**Recall the problem when you kept prompting the user to withdraw money from the wallet to start the game. Integrate this program with the one you have written for problem 7.**

a. For this problem, consider that the minimum balance to start the game is Rs.500.

b. After one game is over add/deduct money from the account.

c. Prompt the user if he/she would like to play one more time.

    i. If the user says “Yes” then check if the user has enough balance to play the game again. If not prompt the user to   withdraw money from his wallet before the program starts the game. When the balance will be greater than Rs. 500 the program will automatically start the game.

    ii. If the user says “No”, then stop the game with a message “Have a wonderful day!”

```python
def Balance_Amount(balance_amt, e_wallet, amount, amount_withdrawn):
    e_wallet = 10000
    while True:
        if balance_amt < 500:
            print("You don't have enough money to start playing.....")
            print("Please withdraw the amount to play...")
            amount = int(input("Enter the amount to withdraw: "))
            amount_withdrawn = amount_withdrawn + amount
            balance_amt = balance_amt + amount
            print("You have balance Rs.", balance_amt)
            e_wallet = e_wallet - amount
            print("You have e-wallet balance of Rs.", e_wallet)

            if balance_amt >= 500:
                print("Welcome! Now, you can start the game.......")
                break
        elif balance_amt >= 500:
            print("Welcome, You can start the game now.....")
            break
    return(balance_amt, e_wallet, amount, amount_withdrawn)
```
